```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

plt.scatter(x,y)
plt.show()
```

```python
# 기울기 a와 절편 b 초기화
a = 0
b = 0

# 학습률
lr = 0.03

# 몇번 반복할지
epochs = 2001 
```

```python
n = len(x)

for i in range(epochs):
    y_pred = a * x +b
    error = y- y_pred
    
    a_diff = (2/n) * sum(-x*(error))
    b_diff = (2/n) * sum(-(error))
    
    a = a -lr * a_diff
    b = b -lr * b_diff
    
    if i %100 ==0:
        print('epochs=%.f, 기울기=%.4f, 절편=%.4f'%(i,a,b))
        
```

```python
y_pred = a * x+ b

plt.scatter(x,y)
plt.plot(x,y_pred,'r')
plt.show()
```

### 다중선형회귀

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.array([2,4,6,8]) # x1 : 공부한 시간
x2 = np.array([0,4,2,3]) # x2 : 과외 시간
y = np.array([81,93,91,97]) # y : 실제 받은 성적
```

```python
# 데이터 분포 확인

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter3D(x1,x2,y)
plt.show()
```

```python
# 변수 설정 

w1 = 0
w2 = 0
b = 0

# 학습률 설정 (임의로 정한다)
lr = 0.01

# 몇번 반복할지 설정
epochs = 4001
```

```python
# 경사하강법

n = len(x1)

for i in range(epochs) : 
    y_pred = w1 * x1 + w2 * x2 +b
    error = y - y_pred
    
    # 편미분
    w1_diff = (2/n) * sum(-x1 * (error))
    w2_diff = (2/n) * sum(-x2 * (error))
    b_diff = (2/n) * sum(-(error))
    
    w1 = w1 - lr * w1_diff
    w2 = w2 - lr * w2_diff
    b = b - lr * b_diff
    
    if i % 100 ==0 : 
        print('epoch : %.f, 기울기 w1 : %.4f, 기울기 w2 : %.4f, 절편:%.4f'%(i,w1,w2,b))
```

```python
# 실제 점수와 예측 점수 출력

print('실제점수 : ',y)
print('예측점수: ',y_pred)
```

### 텐서플로우 활용

```python
# 텐서플로우 활용, 선형회귀모델 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])
```

```python
model = Sequential()

# 출력값, 입력 변수, 분석 방법 (활성화 함수) 설정 
model.add(Dense(1, input_dim = 1, activation = 'linear')) # 출력값 하나 입력값 하나 

# 오차 수정을 위해 경사하강법 (sgd) 사용
# 오차의 정도를 판단하기 위한 평가방법 >> mse(평균제곱오차) 사용 

model.compile(optimizer='sgd', loss='mse')

# 오차를 최소화하는 과정을 2000번 반복
model.fit(x,y,epochs=2001)
```

```python
# 모델 테스트 

hour = 7


prediction = model.predict([hour])

print('%.f 시간을 공부하면, 예상점수는 %.02f 점입니다.' %(hour,prediction))
```

```python
x = np.array([[2,0],[4,4],[6,2],[8,3]]) # 2차원 
y = np.array([81,93,91,97])
```

```python
model = Sequential()

model.add(Dense(1,input_dim=2,activation='linear')) # 2차원 
model.compile(optimizer='sgd', loss='mse')

model.fit(x,y, epochs=2001)
```

```python
# 모델 테스트 

hour = 7
private_class = 4

prediction = model.predict([[hour,private_class]]) # 2차원 

print('%.f 시간을 공부하고 %.f 시간의 과외를 받을 경우, 예상점수는 %.02f 점입니다.' %(hour,private_class, prediction))
```
