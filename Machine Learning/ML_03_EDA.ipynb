```python
# CREDIT 데이터

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import os
```

```python
# 데이터 불러오기 

path = './data/Credit.csv'

credit = pd.read_csv(path)
credit[:10]
```

```python
# 데이터 타입 확인 

credit.dtypes
```

```python
# 데이터 통계 확인 

credit.info()
```

```python
credit.describe()
```

```python
# - mean / median(50%) : rating 비슷 / income, limit 차이가 큼 
# - 대부분 변수의 표준편차가 큼 
# min-max : 데이터 범주가 넓은 편임
```

```python
credit.boxplot()
```

```python
# scatter plot, histogram 확인 

g = sns.PairGrid(credit)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# 더미변수 

#'Student' >> 문자 (YES, NO) >> 숫자(0,1) 변경 >> Student_dummy 

credit['Student']
credit['Student_dummy'] = credit['Student'].replace(to_replace=['Yes', 'No'], value= [0,1])
```

```python
credit[:10]
```

```python
credit.describe(include='all')
```

```python
# 관계성 (상관관계 여부)

# balance (통장잔고) / income(수입)/ student(학생 유무) 관계성 알아보기 

# balance와 income 관계

fig, ax = plt.subplots()

# 조건: 학생 여부
condition = (credit['Student'] == 'Yes')

income_student = credit[condition]

condition = (credit['Student'] == 'No')

income_non_student = credit[condition]

ax.scatter(income_student['Income'], income_student['Balance'],
           s = 10, marker='^', c='orange')

ax.scatter(income_non_student['Income'], income_non_student['Balance'],
           s = 10, marker='s', c='blue')

ax.set(xlabel=('Income'), 
       ylabel=('Balance'))

plt.grid()
plt.show()
```

```python
sns.boxplot(x=credit['Student'], y=credit['Balance'], data = credit)
```

```python
# 데이터 타입 변경 
# horsepower 의 데이터 타입 (object) >> numeric 변경 

# pd.to_numeric(auto['horsepower'])
# ValueError: Unable to parse string "?" at position 32
```

```python
auto['horsepower'].head(40)
```

```python
auto['horsepower'] = auto['horsepower'].replace(to_replace='?', value=0)
```

```python
auto['horsepower'].head(40)
```

```python
auto['horsepower'] = pd.to_numeric(auto['horsepower'])
auto.dtypes
```

```python
auto['horsepower'] = auto['horsepower'].astype('int')
auto['horsepower'].dtype
```

```python
# mpg와 horsepower 관계성 알아보기 

g = sns.PairGrid(auto, y_vars=['mpg'], x_vars=['horsepower'], height=4)
g.map(sns.regplot, color='.3')
```
