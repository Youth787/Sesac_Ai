```python
# auto data 활용, 다항회귀분석

path = '/content/drive/MyDrive/새싹sw-인공지능 서비스 개발자/CLASSROOM/CODE PDF/머신러닝/Auto.csv'
auto = pd.read_csv(path)
```

```python
auto
```

```python
# horsepower ? 값을 0 대체 

auto['horsepower'][:40]
```

```python
auto['horsepower']= auto['horsepower'].replace(to_replace='?',value=0)
```

```python
auto.dtypes
```

```python
# horsepower의 데이터타입 변경 (object >> numeric 숫자)

auto['horsepower'] = pd.to_numeric(auto['horsepower'])
```

```python
auto['horsepower'].dtype
```

```python
# auto data 다중선형회귀

# horsepower의 다양한 feature 생성

auto['horsepower_2'] = auto['horsepower']**2
auto['horsepower_3'] = auto['horsepower']**3
auto['horsepower_4'] = auto['horsepower']**4
auto['horsepower_5'] = auto['horsepower']**5
auto.head(5)
```

```python
# train/test data 분리 

train= auto[:-40]
test = auto[-40:]

print(train)
print(test)
```

```python
# 훈련용 데이터의 feature와 response 분리

auto.columns
```

```python
train_x =\
train[['horsepower','horsepower_2','horsepower_3', 'horsepower_4', 'horsepower_5']]
```

```python
train_y = train[['mpg']]
```

```python
test_x = test[['horsepower','horsepower_2','horsepower_3', 'horsepower_4', 'horsepower_5']]
```

```python
test_y = test[['mpg']]
```

```python
# 선형회귀 객체 생성

regr = linear_model.LinearRegression()
```

```python
# 훈련용 데이터 이용 >> 모델 적합하게 한다. 

regr.fit(train_x[['horsepower','horsepower_2','horsepower_3']],train_y)
```

```python
# 훈련용 데이터 이용, 예측

train_y_pred = regr.predict(train_x[['horsepower','horsepower_2','horsepower_3']])
```

```python
# 테스트 데이터 이용, 예측 

test_y_pred = regr.predict(test_x[['horsepower','horsepower_2','horsepower_3']])
```

```python
# coeffients

print('cofficents', regr.coef_)
```

```python
# train data MSE

print('Train_data mse: %.3f'%mean_squared_error(train_y,train_y_pred))
```

```python
# train data MSE

print('Train_data mse: %.3f'%mean_squared_error(test_y,test_y_pred))
```

```python
# train data r2_score

print('r2_score: %.3f'%r2_score(train_y,train_y_pred))
```

```python
# test data r2_score

print('r2_score: %.3f'%r2_score(test_y,test_y_pred))
```
