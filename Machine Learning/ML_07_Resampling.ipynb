```python
# 재표본 추출 방법 (resampling method)

import numpy as np
import pandas as pd

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Resampling 관련 라이브러리

from sklearn.model_selection import KFold # k- fold 교차검증
from sklearn.utils import resample # bootstrap

# 임의로 데이터를 섞기 위한 라이브러리 (random shuffling)
# 편향 (bias) 방지하기 위해서

from sklearn.utils import shuffle

# 붓꽃(iris) 데이터 로딩 
from sklearn.datasets import load_iris
```

```python
# K-fold 맛보기 
X = ['a','b','c','d']
kf = KFold(n_splits=2) # 2개로 나누기 

for train, test in kf.split(X):
  print("%s %s"%(train, test))
  print(X[train[0]], X[train[1]],X[test[0]], X[test[1]])
```

```python
# 붓꽃 (iris) 데이터 예제_ k-fold cross validation
```

```python
X_iris, y_iris = load_iris(return_X_y = True) 
```

```python
# 데이터를 임의로 섞기 

X_iris, y_iris= shuffle(X_iris,y_iris)
```

```python
X_iris
```

```python
y_iris
```

```python
# 5 fold cross validation

kf = KFold(n_splits=5)

idx = 1

for train, val in kf.split(X_iris):
  print("Fold: #", idx)

  # k-fold 결과 확인 
  print('%s %s'%(train, val))
  print(y_iris[train])
  print(y_iris[val])

  print('----------')
  idx+=1
  
```

```python
# 5 fold cross validation

kf = KFold(n_splits=5)

idx = 1

for train, val in kf.split(X_iris):
  print("Fold: #", idx)

  # k-fold 결과 확인 
  # print('%s %s'%(train, val))
  # print(y_iris[train])
  # print(y_iris[val])

  # LDA (분류 알고리즘)
  # 훈련 (train data로만 훈련)
  clf = LinearDiscriminantAnalysis().fit(X_iris[train],y_iris[train])

  # 클래스 예측
  print('true label: ',y_iris[val])
  print('predicted label: ', clf.predict(X_iris[val]))

  # 정확도(accuracy)
  print('정확도(accuracy): ',clf.score(X_iris[val], y_iris[val]))


  print('----------')
  idx+=1
```

```python
# bootstrap 맛보기

X = np.array([[1.,0.],[2.,1.],[0.,0.]])
y = np.array([0,1,2])

X,y = resample(X,y)
```
