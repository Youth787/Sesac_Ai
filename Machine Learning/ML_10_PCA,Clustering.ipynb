```python
# PCA 주요인 분석

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# pca library 불러오기
from sklearn.decomposition import PCA

# clustering library 불러오기 
from sklearn.cluster import KMeans

# iris data 불러오기
from sklearn.datasets import load_iris
```

```python
from numpy.matrixlib.defmatrix import N
# PCA 간단한 예제 맛보기

X = np.array([[-1,-1],[-2,-1],[-3,-2],[1,1],[2,1],[3,2]])
pca = PCA(n_components=2)
# n_components=2 >> 원하는 개수만큼 pc 를 추출해 냄 
```

```python
# 적합 (fit)
pca.fit(X)
```

```python
X
```

```python
# feature space에서 데이터 확인 

plt.scatter(X[:,0],X[:,1])
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid()
plt.show()
```

```python
# PC score (비지도 학습) >> fit_transform()

Z = pca.fit_transform(X)
```

```python
# feature space에서 데이터 확인 

plt.scatter(Z[:,0],Z[:,1])
plt.xlabel('Z1')
plt.ylabel('Z2')
plt.grid()
plt.show()
```

```python
# Z1축이 Z2축보다 더 큰 variance를 가지고 있다. 
```

```python
# PC loading vector

pca.components_
```

```python
phi = pca.components_
```

```python
# Z

np.matmul(X,phi.T)
```

```python
# PVE(proportion of varience explained) 설명가능한 분산 비율 출력 

pca.explained_variance_ratio_

# 결과해석 :  PC1이 99,2% 설명, PC2: 0.7% 설명명
```

```python
# PCA 예제 (iris data)

iris = load_iris()
```

```python
# iris data를 feature와 response 분리 

X_iris, y_iris = iris.data, iris.target
```

```python
X_iris
```

```python
pca = PCA(n_components =3 )
```

```python
# 적합
pca.fit(X_iris)
```

```python
# PC score
Z = pca.fit_transform(X_iris)
Z
```

```python
# PC space 에서 데이터 분포 확인

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')
ax.scatter(Z[:,0],Z[:,1],Z[:,2],c=y_iris, s=15)
ax.set_xlabel('Z1')
ax.set_ylabel('Z2')
ax.set_zlabel('Z3')
plt.show()
```

```python
# PVE
pca.explained_variance_ratio_
```

```python
# k=means clustering 간단한 예제

X = np.array([[1,2],[1,4],[1,0],[10,2],[10,4],[10,0]])
```

```python
ax.scatter(X[:,0],X[:,1])
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
plt.show()
```

```python
# 적합

kmeans = KMeans(n_clusters=2,random_state=0).fit(X)
```

```python
# 각각의 데이터가 어느 클러스터에 속하는지 표시시
kmeans.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
